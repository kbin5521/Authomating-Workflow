```python
def all_info(domain):
    whoinfo = whois_info(domain)
    ipinfo = ip(domain)
    hostinfo = host(domain)[0]
    res = response(domain)
    url = res.url
    redirect = check_redirection(domain)
    
    response_status = res.status_code
    domain_name = whoinfo.get('domain_name', 'N/A')
    registrar = whoinfo.get('registrar', 'N/A')
    whois_server = whoinfo.get('whois_server', 'N/A')
    creation_date = whoinfo.get('creation_date', 'N/A')
    expiration_date = whoinfo.get('expiration_date', 'N/A')
    Name_server = whoinfo.get('name_servers', 'N/A')
    registrar_email = whoinfo.get('emails', 'N/A')
    registrant_name = whoinfo.get('name', 'N/A')
    registrant_org = whoinfo.get('org', 'N/A')
    registrant_address = whoinfo.get('address', 'N/A')
    registrant_city = whoinfo.get('city', 'N/A')
    registrant_state = whoinfo.get('state', 'N/A')
    registrant_zipcode = whoinfo.get('zipcode', 'N/A')
    registrant_country = whoinfo.get('country', 'N/A')
    
    dic[domain] = {"URL" : url,
                   "Domain": domain_name,
                   "Registrar" : registrar,
                   "Host" : hostinfo,
                   "Registrant_info" : {'Registrant_name' : registrant_name,
                                        'Registrant_org': registrant_org,
                                        'Registrant_address': registrant_address,
                                        'Registrant_city': registrant_city,
                                        'Registrant_state':registrant_state,
                                        'Registrant_zipcode': registrant_zipcode,
                                        'Registrant_coutnry': registrant_country},
                   "Status" : response_status,
                   "Name_server" : Name_server,
                   "Domain Creation Date" : creation_date,
                   "Domain Expiration Date" : expiration_date,
                   "Redirection" : redirect
                        }
```
